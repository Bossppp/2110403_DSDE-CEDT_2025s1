## What is the id of username "excitedPie4" ?

```python
size = rd.dbsize()
cursor = 0
cursor, keys = rd.scan(cursor=cursor, match='user:*', count=size)

key_list = set()
for x in range(len(keys)):
    key_list.add(keys[x].split(':')[1])

for key in key_list:
    data = rd.get("user:" + key + ":name")
    if data == 'excitedPie4':
        ans = key
        break
print(ans)
```

## How many users that "excitedPie4" follows ?

```python
user_follow = rd.smembers('user:' + ans +':follows')
len(user_follow)
```

## What is the average number of follows per user?

## How many users follows between 5-10 users?

```python
cnt = 0
for key in key_list:
    user_follow_cnt_each = rd.scard('user:' + key + ':follows')
    if (5 <= user_follow_cnt_each) and (user_follow_cnt_each <= 10):
        cnt += 1
print(cnt)
```
